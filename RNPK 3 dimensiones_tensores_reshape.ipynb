```python
import numpy as np
```

```python
x = np.array(42)
print(x.shape)
x.ndim
```

```python
x = np.array([42,3,5,3,6,7])
print(x.shape)
x.ndim
```

```python
# matriz
x = np.array([[42,3,5,3,6,7],
              [66,54,5,3,6,7]])
print(x.shape)
x.ndim
```

```python
# tensor
x = np.array([[[42,3,5,3,6,7],
              [66,54,5,3,6,7]],
              [[42,3,5,3,6,7],
              [66,54,5,3,6,7]],
              [[42,3,5,3,6,7],
              [66,54,5,3,6,7]]])
print(x.shape)
x.ndim
```
